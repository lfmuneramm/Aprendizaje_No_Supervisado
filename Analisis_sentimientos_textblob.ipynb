## Limpieza de datos

```python
def clean_text(text):
  text = re.sub(r'^RT[\s]+', '', text)
  text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)
  text = re.sub(r'#', '', text)
  text = re.sub(r'@[A-Za-z0-9]+', '', text)
  return text
```

```python
df['clean_text'] = df['eng_text'].apply(clean_text)
```

```python
df['clean_text']
```

## Eliminamos los emoticones, pictogramas y dem√°s s√≠mbolos

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticones
                           u"\U0001F300-\U0001F5FF"  # simbolos & pictogramas
                           u"\U0001F680-\U0001F6FF"  # simbolos de mapas y transporte
                           u"\U0001F1E0-\U0001F1FF"  # banderas (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['clean_text'] = df['clean_text'].apply(remove_emojis)
```

# Polaridad de los mensajes

```python
def get_polarity(text):
  analysis = TextBlob(text).sentiment
  return analysis.polarity
```

```python
df['polarity'] = df['clean_text'].apply(get_polarity)
```

```python
df['polarity'].head(3)
```

```python
df[df.polarity < 0.3][['text', 'polarity']].head()
```

```python
def x_range(x):
  if x > 0:
    return 1
  elif x == 0:
    return 0
  else:
    return -1
```

```python
df['result'] = df['polarity'].apply(x_range)
```

```python
df.head(3)
```

```python
df['result'].value_counts()
```

```python
import matplotlib.pyplot as plt
labels = 'Neutro', 'Positivo', 'Negativo'
colors = ['#29A0B1','#167D7F','#98D7C2']
plt.pie(df['result'].value_counts(), labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title('Javascript - Sentimientos', fontsize = 20)
plt.axis('equal')
plt.show()
```
